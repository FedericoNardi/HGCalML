```python
index = 1
poy = data['hit_y'].array()[index]
pox = data['hit_x'].array()[index]
poz = data['hit_z'].array()[index]
ene = data['recHit_dE'].array()[index]
 

'''
plot3d function, showing the colorbar on the side
'''
def plot3d(pox,poy,poz,ene):
    import plotly.graph_objects as go
    import numpy as np
    from plotly.subplots import make_subplots
    fig = make_subplots(rows=1, cols=1, specs=[[{'is_3d': True}]])
    fig.add_trace(go.Scatter3d(x=pox, y=poy, z=poz, mode='markers', marker=dict(
        size=4, 
        symbol='square', 
        color=np.log10(ene), 
        colorscale='Viridis', 
        opacity=0.5,
        colorbar=dict(thickness=20, title='log<sub>10</sub>(Energy [GeV])'),
        )), 1, 1)
    fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
    fig.update_traces(marker=dict(line=dict(width=0)))
    fig.show()


plot3d(pox,poy,poz,ene)
```
